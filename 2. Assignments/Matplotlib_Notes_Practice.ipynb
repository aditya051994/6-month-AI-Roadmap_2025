```python
import numpy as np
x=np.arange(0,10)
y=np.sin(x)
plt.plot(x,y)
plt.xticks([0,2,4,6,8],['Zero','Two','Four','Six','Eight'] )
plt.yticks([1,0,-1],['Max','Zero','Min'] )
plt.show()
```

```python
import numpy as np
x=np.arange(0,10)
y=np.sin(x)
fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
#Legend
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.legend(['Some value'])
plt.show()
```

```python
#Line Plot
x=np.array([1,2,3,4,5,6])
y=x*2
plt.plot(x,y)
plt.show()
```

```python
#Bar chart
data=[.5,25.,50.,20.]
plt.bar(range(len(data)),data,color='r')
```

```python
x=np.random.normal(1,100,10)
plt.hist(x)
```

```python
#Pie Chart
x=[1,2,3,4]
y=['A','B','C','D']
plt.pie(x,labels=y)
plt.show()
```

```python
value=np.random.normal(100,20,300)
plt.boxplot(value,patch_artist=True)
plt.show()
```
